```python
from keras.layers import TextVectorization
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['the glass of milk',
     'the glass of juice',
     'the cup of tea',
     'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
sent
```

```python
voc_size=10000
```

```python
from keras.layers import TextVectorization

vectorizer = TextVectorization(
    max_tokens=voc_size,
    output_mode="int",
    output_sequence_length=10      #max_len in one_hot tf
#it deefault have post padding we cant to pre
)

vectorizer.adapt(sent)
encoded = vectorizer(sent)
encoded
```

```python
from keras.layers import Embedding
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
```

```python
#Doing pre padding

from keras.preprocessing.sequence import pad_sequences

encoded_pre = pad_sequences(encoded, maxlen=10, padding='pre')
encoded_pre
```

```python
dim = 10          # Embedding vector size
voc_sizvoc_size = len(vectorizer.get_vocabulary())  # automatically includes padding token
max_len = 10      # Padded sequence length

model = Sequential()
model.add(Embedding(
    input_dim=voc_size, 
    output_dim=dim, 
    input_length=max_len
))

model.compile('adam','mse')
```

```python
model.summary()
```

```python
print(model.predict(encoded))
```
